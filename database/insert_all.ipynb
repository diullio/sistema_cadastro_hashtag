```python
import os
import sqlite3
import csv
```

```python
# Caminho do banco de dados
database = 'database.db'
```

```python
# Função para cadastrar um produto
def query_cadastrar(dictCadastro):
    try:
        conn = sqlite3.connect(database)
        c = conn.cursor()

        # Inserir dados na tabela
        c.execute('''
            INSERT INTO produtos (cd, marca, tipo, cat, preco_un, custo, obs)
            VALUES (?, ?, ?, ?, ?, ?, ?)
        ''', (dictCadastro['cd'], dictCadastro['marca'], dictCadastro['tipo'], dictCadastro['cat'], dictCadastro['preco_un'], dictCadastro['custo'], dictCadastro['obs']))

        conn.commit()
        return 'sucesso'
    except Exception as e:
        print(f'Ocorreu uma exceção: {e}')
        return None
    finally:
        if conn:
            conn.close()
```

```python
def inserir_csv_no_banco(caminho_csv):
    with open(caminho_csv, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            # Ajustar nomes das chaves para corresponder aos usados na função query_cadastrar
            dictCadastro = {
                'cd': row['codigo'],
                'marca': row['marca'],
                'tipo': row['tipo'],
                'cat': row['categoria'],
                'preco_un': row['preco_unitario'],
                'custo': row['custo'],
                'obs': row['obs'] if 'obs' in row and row['obs'] else None
            }
            resultado = query_cadastrar(dictCadastro)
            if resultado == 'sucesso':
                print(f"Produto {row['codigo']} cadastrado com sucesso.")
            else:
                print(f"Falha ao cadastrar o produto {row['codigo']}.")
```

```python
caminho_csv = 'produtos.csv'
inserir_csv_no_banco(caminho_csv)
```
